# Multicolinearity - VIF

```python
dataset.drop("sl_no",inplace=True,axis=1)
```

```python
dataset
```

```python
import seaborn as sns
```

```python
sns.pairplot(dataset)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(X):
    #calculating VIF
    vif=pd.DataFrame()
    vif["variables"]=X.columns
    vif["VIF"]=[variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
    return(vif)
```

```python
calc_vif(dataset[["ssc_p","hsc_p","degree_p","etest_p","mba_p","salary"]])
```

```python
calc_vif(dataset[["etest_p","salary"]])
```

#Independant Sample-Unpair T Test

Differrent Group(Male,Female) but same contion(salary)

```python
from scipy.stats import ttest_ind
dataset=dataset.dropna()
male=dataset[dataset["gender"]=="M"]["salary"]
female=dataset[dataset["gender"]=="F"]["salary"]
#print(male)
ttest_ind(male,female)
```

Dependent Sample-Paired T_Test

Same Group(Male) but Different Condition(ssc_p,hsc_p)

```python
from scipy.stats import ttest_rel
#dataset=dataset.dropna()
male=dataset[dataset["gender"]=="M"]["ssc_p"]
fmale=dataset[dataset["gender"]=="M"]["hsc_p"]
#print(male)
ttest_rel(male,fmale)
```

Hyphothesis test
Statement:Test whether SSC_P of male and HSC_P of male is same?
Null hyphothesis=There is no silmilarity between SSC_P of male and HSC_P of male

```python
from scipy.stats import ttest_rel
#dataset=dataset.dropna()
male=dataset[dataset["gender"]=="M"]["ssc_p"]
fmale=dataset[dataset["gender"]=="M"]["hsc_p"]
#print(male)
ttest_rel(male,fmale)
```

Reject Null hypothesis p<0.05
Accept Null hypothesis and Reject Alternate hypothesis

ANAVO:Analysis of Variance-one way

```python
import scipy.stats as stats

stats.f_oneway(dataset["ssc_p"],dataset["hsc_p"],dataset["degree_p"])
```

ANAVO:Analysis of Variance-Two-Way

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('salary ~ ssc_p + hsc_p + ssc_p:hsc_p', data=dataset).fit()

anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)
```
